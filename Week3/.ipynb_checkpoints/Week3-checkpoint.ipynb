```python
# Read the PylintOutput file
pylintOutput = open("/Users/apple/Desktop/SummerResearch/Week1/pylintOutput.txt", "a")

# Data Preprocess Stage
pylintOutput.write("************")
pylintOutput.close()

# Store Each line of the Pylint Outcome
pylintOutput = open("/Users/apple/Desktop/SummerResearch/Week1/pylintOutput.txt")
lines = []
while True:
    line = pylintOutput.readline()
    lines.append(line)
    if ("" == line):
        break

# Categorize each student's warning message 
students = []
student = []
for line in lines:
    if line[0:12] == "************":
        students.append(student)
        student = []
    student.append(line)
students.pop(0)

# Process each student
# single_outcome = ['Student_ID', 'error code 1', 'error code 2', ...]
# outcome = [single_outcome1, single_outcome2, ...]

codeReference = set()
outcome = []
for student in students:
    single_outcome = []
    # Get the Student ID first
    indexOfStudentID = student[0].find("Module") + 17
    single_outcome.append(student[0][indexOfStudentID : ].strip('\n'))
    single_outcome.insert(0, int(student[0][indexOfStudentID + 1 : ].strip('\n')))
    
    # Get the error code of this student
    messages = student[1 : ]
    for message in messages:
        tokens = message.split() 
        # Pick the warning code
        for token in tokens:
            if (len(token) == 6) and (token[0].isalpha()) and (token[1 : 4].isdigit()) and token[-1] == ':':
                single_outcome.append(token[0 : 5])
                codeReference.add(message[message.find(token[0 : 5]):].strip('\n'))
    outcome.append(single_outcome)

# Sort elements based on students' ID
outcome.sort(key = lambda x:x[0])
for e in outcome:
    e.pop(0)
    
# Find and sort all unique warning code alphabetically 
allErrorCodeSet = set()
# studentInfo = ['Student_ID', 'error code 1', 'error code 2', ....]
for studentInfo in outcome:
    for i in range(1, len(studentInfo)):
        allErrorCodeSet.add(studentInfo[i])
        
allErrorCodes = list(allErrorCodeSet)
allErrorCodes.sort()

# Count the number of each warning code for each student
# processedStudent = [Student_ID, {"c0001": 2, "w0001": 3, ...}]
# processedStudents = [processedStudent1, processedStudent2, ...]

processedStudents = []
for single_outcome in outcome:
    processedStudent = [single_outcome[0]]
    frequencyMap = {}
    for i in range(1, len(single_outcome)):
        if single_outcome[i] not in frequencyMap:
            frequencyMap[single_outcome[i]] = 1
        else:
            frequencyMap[single_outcome[i]] += 1
    processedStudent.append(frequencyMap)
    processedStudents.append(processedStudent)
```

```python
# Get reference to the meaning of each warning code
simpleCodeReference = []
allErrorCodesCpy = list(allErrorCodes)
for ref in codeReference:
    singleSimpleCodeReference = []
    singleSimpleCodeReference.append(ref[0:5])
    start = ref.rfind('(') + 1
    end = ref.rfind(')')
    if start == 0 or end == -1:
        singleSimpleCodeReference.append(ref[8:])
    else:
        singleSimpleCodeReference.append(ref[start : end])
    if singleSimpleCodeReference[0] in allErrorCodesCpy: 
        simpleCodeReference.append(singleSimpleCodeReference)
        allErrorCodesCpy.remove(singleSimpleCodeReference[0])

# Sort error codes
simpleCodeReference.sort(key = lambda x:x[0])
```

```python
# Read the Hyperstyle JSON file
# import required module
from pathlib import Path
import json
 
# assign directory
directory = '/Users/apple/Desktop/SummerResearch/Week3/JSON_output'
 
processedStudentsHyperstyle = []
errorCodeSet = set()
errorCode = []

# iterate over files in that directory
files = Path(directory).glob('*')
for file in files:
    sFile = str(file)
    if sFile[-4:] == "json":
        f = open(file)
        student = json.load(f)
        processedStudent = []
        # Append the name of the solution
        processedStudent.append((int)(sFile[str(file).find("-s") + 2 : sFile.find(".")]))
        processedStudent.append(sFile[str(file).find("-s") + 1 : sFile.find(".")])
        
        # Append the code quality and text quality
        processedStudent.append(student["quality"]["code"])
        processedStudent.append(student["quality"]["text"][student["quality"]["text"].rfind(": ") + 2 : ])
        
        # Collect the student error
        frequencyMap = {}
        for issue in student["issues"]:
            if issue["code"] not in frequencyMap:
                frequencyMap[issue["code"]] = 1
            else:
                frequencyMap[issue["code"]] += 1
        processedStudent.append(frequencyMap)
        processedStudentsHyperstyle.append(processedStudent)
        
        # Collect error codes
        for issue in student["issues"]:
            if issue["code"] not in errorCodeSet:
                errorCodeSet.add(issue["code"])
                singleErrorCode = [issue["code"], issue["text"], issue["category"], issue["difficulty"]]
                errorCode.append(singleErrorCode)
                
# Sort the outcome
processedStudentsHyperstyle.sort(key = lambda x:x[0])
for e in processedStudentsHyperstyle:
    e.pop(0)
# print(processedStudentsHyperstyle)
# print(errorCode)
```

```python
# Merge students' error message from Pylint output and HyperStyle output
mergedInfo = []
sameError = {}
pylintErrorCode = set()
for i in range(len(processedStudentsHyperstyle)):
    singleMergedInfo = [processedStudentsHyperstyle[i][0], processedStudentsHyperstyle[i][1], processedStudentsHyperstyle[i][2], processedStudents[i][1]]
    newIssue = {}
    hyperstyleError = dict(processedStudentsHyperstyle[i][3])
    for issue in processedStudentsHyperstyle[i][3]:
        if issue in singleMergedInfo[3]:
            pylintErrorCode.add(issue)
            print("Same Error:",singleMergedInfo[0], "\t", issue, "\t\t", processedStudentsHyperstyle[i][3][issue], "\t", singleMergedInfo[3][issue])
            if singleMergedInfo[0] not in sameError:
                sameError.update({singleMergedInfo[0]: singleMergedInfo[3][issue]})
            else:
                sameError.update({singleMergedInfo[0]: sameError[singleMergedInfo[0]] + singleMergedInfo[3][issue]})
            if processedStudentsHyperstyle[i][3][issue] != singleMergedInfo[3][issue]:
                print("Not Equal")
            else:
                hyperstyleError.pop(issue)
    singleMergedInfo.append(hyperstyleError)
    singleMergedInfo.append((int)(singleMergedInfo[0][1:]))
    mergedInfo.append(singleMergedInfo)

# Sort the outcome
mergedInfo.sort(key = lambda x:x[-1])
for e in mergedInfo:
    e.pop(-1)

# print(mergedInfo) 
# print(pylintErrorCode)
```

```python
# Add some statistics into mergeInfo 
for i in range(len(mergedInfo)):
    pylintNum = 0
    for num in mergedInfo[i][3]:
        pylintNum += mergedInfo[i][3][num]
    mergedInfo[i].append(pylintNum)
    pylintNum = 0
    for num in mergedInfo[i][4]:
        pylintNum += mergedInfo[i][4][num]
    if mergedInfo[i][0] in sameError:
        pylintNum +=sameError[mergedInfo[i][0]]
    mergedInfo[i].append(pylintNum)
# print(mergedInfo)
```

```python
# Collect Pylint Error Code in common with hyperstyle
pylintCode = []
hyperCode = []
for singleCode in errorCode:
    if singleCode[0] in pylintErrorCode:
        pylintCode.append(singleCode)
    else: # Collect Hyperstyle Error Code
        hyperCode.append(singleCode)
        
# Sort error codes
pylintCode.sort(key = lambda x:x[0])

# Sort error codes
hyperCode.sort(key = lambda x:x[0])
               
# print(pylintCode)
# print(hyperCode)
```

```python
# import package
import xlsxwriter 
from openpyxl import Workbook
from openpyxl.styles import Alignment
from openpyxl.styles import PatternFill
from openpyxl.styles import Font

# Create Workbook 
workbook = Workbook()

# Delete the empty worksheet
del workbook["Sheet"]

# Sheets Names
sName = "Comparision of Tools' Outcomes"

# Create Sheet
workbook.create_sheet(sName)

# Choose worksheet
sheet = workbook[sName]
```

```python
# Create worksheet x-axis title
sheet["A1"] = "Student ID"
sheet.column_dimensions['A'].width = 30
sheet["A1"].alignment = Alignment(horizontal = 'center')
index = 0
for row in sheet.iter_rows(min_row = 1, min_col = 2, max_row = 1, max_col = len(simpleCodeReference) + 1):
    for cell in row:
        sheet.column_dimensions[xlsxwriter.utility.xl_col_to_name(cell.column - 1)].width = 30
        cell.value = simpleCodeReference[index][0]
        cell.alignment = Alignment(horizontal = 'center')
        cell.fill = PatternFill(start_color = "00FFFF00", end_color = "00FFFF00", fill_type = "solid")
        cell.font = Font(bold = True)
        index += 1
        
index = 0
for row in sheet.iter_rows(min_row = 1, min_col = len(simpleCodeReference) + 5, max_row = 1, max_col = len(simpleCodeReference) + len(hyperCode) + 4):
    for cell in row:
        sheet.column_dimensions[xlsxwriter.utility.xl_col_to_name(cell.column - 1)].width = 40
        cell.value = hyperCode[index][0]
        cell.alignment = Alignment(horizontal = 'center')
        cell.fill = PatternFill(start_color = "00FFFF00", end_color = "00FFFF00", fill_type = "solid")
        cell.font = Font(bold = True)
        index += 1
```

```python
# For maxErrorComp (color gradient for comparing two tools)
maxErrorComp = 0
for info in mergedInfo:
    bigger = 0
    if info[5] > info[6]:
        bigger = info[5]
    else:
        bigger = info[6]
        
    if maxErrorComp < bigger:
        maxErrorComp = bigger
```

```python
# Fill in the information of each student (Pylint)


# Define the level of numbers of errors (for visualizarion later)
maxError = 0
for processedStudent in processedStudents:
    if maxError < max(processedStudent[1].values()):
        maxError = max(processedStudent[1].values())

studentIndex = 0
for row in sheet.iter_rows(min_row = 2, min_col = 1, max_row = len(processedStudents) + 1, max_col = len(simpleCodeReference) + 1):
    flag = True
    errorCodeIndex = 0
    for cell in row:
        if flag:
            cell.value = processedStudents[studentIndex][0]
            cell.font = Font(bold = True)
            flag = False
        else:
            # Loop through set of error codes
            errorCodes = processedStudents[studentIndex][1]
            if allErrorCodes[errorCodeIndex] in errorCodes:
                cell.value = errorCodes[allErrorCodes[errorCodeIndex]]
                # Add gradient colors for better visualization
                hexColor = '#%02x%02x%02x' % ((int)(255 - ((cell.value / maxError) * 255)), (int)(255 - ((cell.value / maxError) * 155)), 255)
                cell.fill = PatternFill(start_color = str(hexColor)[1:], end_color = str(hexColor)[1:], fill_type = "solid")
            errorCodeIndex += 1  
        cell.alignment = Alignment(horizontal = 'center')
    studentIndex += 1
  

# Count warning message occurrence

# Title of the row
sheet.cell(row = len(processedStudents) + 2, column = 1).value = "Warning Message Occurrence"
sheet.cell(row = len(processedStudents) + 2, column = 1).font = Font(bold = True)
sheet.cell(row = len(processedStudents) + 2, column = 1).alignment = Alignment(horizontal = 'center')

# Fill the frequency of each warning message
errorFreq = []
for singleError in allErrorCodes:
    errorCounter = 0
    for processedStudent in processedStudents:
        if singleError in processedStudent[1]:
            errorCounter += processedStudent[1][singleError]
    errorFreq.append(errorCounter)

maxError = max(errorFreq)
errorIndex = 0
for row in sheet.iter_rows(min_row = len(processedStudents) + 2, min_col = 2, max_row = len(processedStudents) + 2, max_col = len(allErrorCodes) + 1):
    for cell in row:
        cell.value = errorFreq[errorIndex]
        cell.alignment = Alignment(horizontal = 'center')
         # Add gradient colors for better visualization
        hexColor = '#%02x%02x%02x' % (255, (int)(255 - ((cell.value / maxError) * 255)), (int)(255 - ((cell.value / maxError) * 255)))
        cell.fill = PatternFill(start_color = str(hexColor)[1:], end_color = str(hexColor)[1:], fill_type = "solid")
        errorIndex += 1

        
# Fill in the Total Pylint Warning per Student
sheet.cell(row = 1, column = len(allErrorCodes) + 2).value = "Total Pylint Warning per Student"
sheet.cell(row = 1, column = len(allErrorCodes) + 2).font = Font(bold = True)
sheet.cell(row = 1, column = len(allErrorCodes) + 2).fill = PatternFill(start_color = "00FFFF00", end_color = "00FFFF00", fill_type = "solid")
sheet.cell(row = 1, column = len(allErrorCodes) + 2).alignment = Alignment(horizontal = 'center')
sheet.column_dimensions[xlsxwriter.utility.xl_col_to_name(len(allErrorCodes) + 1)].width = 30
studentIndex = 0
# maxError = 0
for row in sheet.iter_rows(min_row = 2, min_col = len(allErrorCodes) + 2, max_row = len(processedStudents) + 1, max_col = len(allErrorCodes) + 2):
    for cell in row:
        cell.value = mergedInfo[studentIndex][5]
#         if maxError < cell.value:
#             maxError = cell.value
        cell.alignment = Alignment(horizontal = 'center')
        # Add gradient colors for better visualization
        hexColor = '#%02x%02x%02x' % (255, (int)(255 - ((cell.value / maxErrorComp) * 255)), (int)(255 - ((cell.value / maxErrorComp) * 255)))
        cell.fill = PatternFill(start_color = str(hexColor)[1:], end_color = str(hexColor)[1:], fill_type = "solid")
        studentIndex += 1
 

# Add code reference to excel
# Title of the row
sheet.cell(row = len(processedStudents) + 3, column = 1).value = "Code Reference"
sheet.cell(row = len(processedStudents) + 3, column = 1).font = Font(bold = True)
sheet.cell(row = len(processedStudents) + 3, column = 1).alignment = Alignment(horizontal = 'center')
refIndex = 0
for row in sheet.iter_rows(min_row = len(processedStudents) + 3, min_col = 2, max_row = len(processedStudents) + 3, max_col = len(allErrorCodes) + 1):
    for cell in row:
        cell.value = simpleCodeReference[refIndex][1]
        cell.alignment = Alignment(horizontal = 'center')
        refIndex += 1
```
